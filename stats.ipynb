```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Analysis of the Ratio of Women’s Wages to Men’s Wages by Region in Kyrgyzstan (2002–2024)

df = pd.read_csv('me.csv', encoding='cp1251')

df_clean = df.set_index('Region')

for col in df_clean.columns:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

print("\nCleaned DataFrame (df_clean) info:")
df_clean.info()
print("\nCleaned DataFrame (df_clean) head:")
print(df_clean.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
for region in df_clean.index:
    plt.plot(df_clean.columns, df_clean.loc[region], marker='o', label=region)
plt.title("Ratio of female to male wages by region (2002–2024)")
plt.xlabel("Year")
plt.ylabel("% of male wages")
plt.legend()
plt.grid(True)
plt.show()
```

The line chart demonstrates an overall upward trend across all regions, meaning that the gender wage gap has been gradually decreasing over time.
In the early 2000s, women earned on average about 65% of men’s wages, while by 2024 this figure increased to 70–80% depending on the region.

```python
plt.figure(figsize=(9, 5))
sns.barplot(x=df_clean.index, y=df_clean['2024'])
plt.title("Ratio of female to male wages by region (2024)")
plt.ylabel("% of male wages")
plt.xticks(rotation=45)
plt.show()
```

The highest ratios are observed in Bishkek and Chui region (around 80–85%),
while the lowest ratios are found in Naryn and Batken regions (around 70%).
This indicates economic inequality among regions — women in more urbanized and economically developed areas tend to earn closer to men’s levels.

```python
periods = {
    "2002–2008": df_clean.loc[:, '2002':'2008'].mean(axis=1),
    "2009–2015": df_clean.loc[:, '2009':'2015'].mean(axis=1),
    "2016–2024": df_clean.loc[:, '2016':'2024'].mean(axis=1),
}
stacked_df = pd.DataFrame(periods)

stacked_df.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Average ​​of female wages (% of male's) over time")
plt.ylabel("%")
plt.legend(title="Period")
plt.xticks(rotation=45)
plt.show()
```

The stacked bars clearly demonstrate a gradual increase in all regions.
The difference between the first and the last periods is about 10–15 percentage points,
which confirms a consistent improvement in wage equality.

```python
plt.figure(figsize=(8, 5))
plt.hist(df_clean.values.flatten(), bins=10, color='purple', edgecolor='black')
plt.title("Distribution of salary ratios(all regions, all years))")
plt.xlabel("% of male wages")
plt.ylabel("Periodicity")
plt.show()
```

Most values are concentrated between 65% and 80%, meaning that in most cases women earn roughly two-thirds of men’s wages.
Only a few values above 85% occur in some regions and years.

```python
plt.figure(figsize=(7, 7))
mean_values = df_clean.mean(axis=1)
plt.pie(mean_values, labels=df_clean.index, autopct='%1.1f%%', startangle=140)
plt.title("Average ratio female and male wages by region (2002-2024)")
plt.show()
```

The largest portion belongs to Bishkek, where women’s wages are closest to men’s (around 82%).
Chui and Osh regions also show relatively high values,
while Naryn and Batken occupy smaller portions, indicating wider pay gaps.

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df_clean.T, orient='h')
plt.title("Distribution of the ratio female and male wages by region(2002–2024)")
plt.xlabel("% of male wages")
plt.show()
```

Regions like Bishkek and Chui have narrower, more stable distributions,
while mountainous regions such as Naryn and Batken show greater variation,
indicating more fluctuation in wages over the years.
